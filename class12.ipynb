# Modes of file

```python
with open("./abc.text",'r') as file: 
    print(file.read())
```

```python
with open("./xyz.text",'w') as file:  # xyz.text pehle se bani howi nhi thi tw us ne khud hi bana jay data write kara dia
    file.write("Cursor start - Add new data by overriding previous")
```

```python
with open("./abc.txt",'w') as file: 
    print(file.read())
    file.write("I wont be able to read now since the mode is 'w' ")
```

```python
with open("./abc1.txt",'r+') as file: 
    print(file.read())
    file.write("Cursor start - in this example read then write")
```

```python
with open("./abc2.txt",'r+') as file: 
    file.write("I will be able to read and write as well since the mode is 'r+'")
    file.seek(0)
    print(file.read())
```

```python
with open("./abc3.txt",'a') as file: 
    file.write("Cursor at the end of the content") #but it will not override as w it will append the data
```

```python
with open("./abc4.txt",'x') as file:  
    file.write("It creates new file but if the file already exists then there is an error") 
```

```python
with open("./abc5.txt",'rb+') as file:  
    print(file.read()) #only used for reading,reads text in binary mode
```

```python
with open("./abc6.txt",'w+') as file: 
    print(file.read())
    file.write("I wont be able to read now since the mode is 'w' ")
```

```python
with open("./abc7.txt",'a+') as file: 
    print(file.read())
    file.write("I wont be able to read now since the mode is 'w' ")
```

```python
from typing import list
import PyPDF2

def read_pdf(file_path : str) -> List[str]:
    with open(file_path,'rb') as file:
        reader : PyPDF2.PdfReader = PyPDF2.PdfReader(file)
        text_content : List[str] = [page.extract_text() for page in reader.pages]
        return text_content

pages : list[str] = read_pdf("pdf link here saved in the smae pc")
pages
```
